```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('Unemployment_Rate_upto_11_2020.csv') #Read dataset
df.sample(10)  #fetch ten sample of dataset
```

```python
import matplotlib.pyplot as plt
```

```python
#visualization

fig=plt.figure()
ax0=fig.add_subplot(1,2,1)
df[:10].plot(kind="bar",color="blue",figsize=(20,10),ax=ax0)
ax0.set_title("Top 10 States with Highest Unemployment")
ax0.set_xlabel("State")
ax0.set_ylabel("Number of people unemployed in %")
```

```python
plt.figure(figsize=(6,6))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

## shows regionwise count of Estimated Unemployment Rate(%)

```python
plt.figure(figsize=(10,10))
plt.title("Unemployment in india")
sns.histplot(x=' Estimated Unemployment Rate (%)',hue= "Region", data=df,kde=False)
plt.show()
```
